# Sales Performance Analysis

```python
# Top-selling products by quantity

top_products_qty = df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
print("Top 10 Products by Quantity:\n", top_products_qty)
```

```python
# Top-selling products by revenue

top_products_rev = df.groupby("Description")["Revenue"].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Products by Revenue:\n", top_products_rev)
```

```python
# Top customers by spending

top_customers = df.groupby("CustomerID")["Revenue"].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Customers by Spending:\n", top_customers)
```

```python
# Best-selling days & times

df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()
```

```python
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
sales_by_day = df.groupby("DayOfWeek")["Revenue"].sum().sort_values(ascending=False)
sales_by_hour = df.groupby("Hour")["Revenue"].sum()
```

```python
print("\nSales by Day of Week:\n", sales_by_day)
```

```python
print("\nSales by Hour:\n", sales_by_hour)
```

```python
# Most profitable product categories (group by keywords)

categories = {
    "HOLDER": ["HOLDER"],
    "SET": ["SET"],
    "CUP": ["CUP", "MUG"],
    "HEART": ["HEART"],
    "LANTERN": ["LANTERN"]
}
```

```python
def assign_category(desc):
    desc = str(desc).upper()
    for cat, keywords in categories.items():
        for kw in keywords:
            if kw in desc:
                return cat
    return "OTHER"
```

```python
df['Category'] = df['Description'].apply(assign_category)
category_revenue = df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
```

```python
print("\nMost Profitable Product Categories:\n", category_revenue)
```

```python
import plotly.express as px

fig = px.bar(top_products_rev, x=top_products_rev.index, y=top_products_rev.values, title="Top Products by Revenue")
fig.show()

fig2 = px.bar(category_revenue, x=category_revenue.index, y=category_revenue.values, title="Most Profitable Product Categories")
fig2.show()
```

# Time-Series & Seasonal Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Data Cleaning

```python
df = pd.readdf = df.dropna(subset=["CustomerID"])
```

```python
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
df = df[df['Quantity'] > 0]
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
# Monthly Sales Trends


df['Month'] = df['InvoiceDate'].dt.to_period('M')
monthly_sales = df.groupby('Month')['Revenue'].sum().to_timestamp()
```

```python
print("Monthly Sales:\n", monthly_sales)
```

```python
# Plot monthly sales trend


plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

```python
# Day-of-week & Time-of-day Patterns


df['DayOfWeek'] = df['InvoiceDate'].dt.day_name()
df['Hour'] = df['InvoiceDate'].dt.hour
```

```python
sales_by_day = df.groupby('DayOfWeek')['Revenue'].sum().reindex(
    ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
)
print("\nSales by Day of Week:\n", sales_by_day)
```

```python
sales_by_hour = df.groupby('Hour')['Revenue'].sum()
print("\nSales by Hour:\n", sales_by_hour)
```

```python
# Plot day-of-week


plt.figure(figsize=(8,4))
sales_by_day.plot(kind='bar', color='skyblue')
plt.title("Revenue by Day of Week")
plt.xlabel("Day")
plt.ylabel("Revenue")
plt.show()
```

```python
# Plot hour-of-day


plt.figure(figsize=(8,4))
sales_by_hour.plot(kind='bar', color='orange')
plt.title("Revenue by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Revenue")
plt.show()
```

```python
# Year-over-Year / Month-over-Month Growth

monthly_sales.index = pd.to_datetime(monthly_sales.index)
```

```python
# MoM growth

mom_growth = monthly_sales.pct_change().dropna() * 100
print("\nMonth-over-Month Growth (%):\n", mom_growth)
```

```python
# YoY growth (12 months)

yoy_growth = monthly_sales.pct_change(12).dropna() * 100
print("\nYear-over-Year Growth (%):\n", yoy_growth)
```

```python
import streamlit as st
import plotly.express as px

st.subheader("Monthly Sales Trend")
fig = px.line(monthly_sales, x=monthly_sales.index, y=monthly_sales.values, title="Monthly Sales Trend")
st.plotly_chart(fig)

st.subheader("Revenue by Day of Week")
fig2 = px.bar(sales_by_day, x=sales_by_day.index, y=sales_by_day.values)
st.plotly_chart(fig2)

st.subheader("Revenue by Hour of Day")
fig3 = px.bar(sales_by_hour, x=sales_by_hour.index, y=sales_by_hour.values)
st.plotly_chart(fig3)
```
